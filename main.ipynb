1.5. Cleaning of categorical data:
- Standardize the values of categorical variables that are strings so that they are all in lowercase. This will allow for the merging of different databases and ensure that strings are not considered distinct due to case sensitivity, with some being in uppercase and others in lowercase.

```python
data1['Category'] = data1['Category'].str.lower()
data1['Item'] = data1['Item'].str.lower()
data1['Serving Size'] = data1['Serving Size'].str.lower()
```

```python
data1[data1.select_dtypes(
    include=['float64', 'int64']).columns] = data1[data1.select_dtypes(
        include=['float64', 'int64']).columns].applymap(
            lambda x: np.nan if x < 0 else x)
```

1.7. Dealing with NaNs:

- Obtain a heatmap where you can easily visualize the NaNs present in the dataset.
- Remove rows and columns with all values set to NaN.
- Calculate the percentage of NaNs present in each variable.
- Eliminate categories and variables (columns) that have more than 60% of NaNs.
- For variables with less than 10% of NaNs, impute the values with their median, calculated using the available data for that variable.


```python
import seaborn as sns

# Obtain a heatmap where you can easily visualize the NaNs present in the dataset.
ax = sns.heatmap(data1.isnull().transpose(), cmap='Blues')
plt.title('Visualization of missing values in the dataframe data1', fontweight='bold', pad=20)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=8, fontweight='bold')
ax.set_yticklabels(ax.get_yticklabels(), fontsize=8, fontweight='bold')
plt.show()
```

```python
# Remove rows and columns with all values set to NaN.
data1.dropna(how='all', axis=0, inplace=True)
data1.dropna(how='all', axis=1, inplace=True)
```

```python
# Calculate the percentage of NaNs present in each variable.
data1.isnull().mean() * 100
```

```python
# Eliminate categories and variables (columns) that have more than 60% of NaNs.
data1 = data1.dropna(thresh=len(data1.columns)*0.6, axis=0)
data1 = data1.loc[:, (data1.isnull().sum() / len(data1) * 100) < 60]
```

```python
# For variables with less than 10% of NaNs, impute the values with their median, calculated using the available data for that variable. 
from sklearn.impute import SimpleImputer

new_percent = (data1.select_dtypes(include=['float64', 'int64']).isnull().sum() / data1.shape[0]) * 100
cols_to_impute = new_percent[new_percent < 10].index.tolist()
imputer = SimpleImputer(strategy='median')
data1[cols_to_impute] = imputer.fit_transform(data1[cols_to_impute])
```

3. Is the distribution of calories statistically different between 'Salads' and 'Desserts'? What about the sugar distribution? In each case (calories and sugar), display the distributions for Salads and Desserts on the same graph to better observe the discrepancies.

```python
merged.columns
```

```python
# Is the distribution of calories statistically different between 'Salads' and 'Desserts'? What about the sugar distribution?

salads = merged[merged['Category_x'] == 'salads']
desserts = merged[merged['Category_x'] == 'desserts']
```

```python
# Display both distributions on the same graph for graphical comparison as well

plt.figure(figsize=(10, 6))

sns.kdeplot(salads['Calories'], label='Salads')
sns.kdeplot(desserts['Calories'], label='Desserts')
plt.xlabel('Calories')
plt.ylabel('Density')
plt.title('Distribution of Calories: Salads vs Desserts')
plt.legend()

sns.despine()  # Remove the top and right axes

plt.tight_layout()
plt.show()
```

```python
# Are the calorie distributions statistically different in 'Salads' and 'Desserts'? What about sugar?

# Student's t-test for two independent samples
from scipy.stats import ttest_ind
salads_calories = salads['Calories']
desserts_calories = desserts['Calories']

salads_sugar = salads['Sugars']
desserts_sugar = desserts['Sugars']

print('|---------------------------------------------------------------------------------------------|')
# Hypothesis for calorie distribution
t_stat_calories, p_val_calories = ttest_ind(salads_calories, desserts_calories)
print('\tCalories:\tt-statistic:', t_stat_calories, '\tp-value:', p_val_calories)

# Hypothesis for sugar distribution
t_stat_sugar, p_val_sugar = ttest_ind(salads_sugar, desserts_sugar)
print('\tSugar:\t\tt-statistic:', t_stat_sugar, '\tp-value:', p_val_sugar)
print('|---------------------------------------------------------------------------------------------|')

if p_val_calories < 0.05:
    print("\tThe calorie distribution is statistically different between Salads and Desserts.")
else:
    print("\tThe calorie distribution is not statistically different between Salads and Desserts.")

if p_val_sugar < 0.05:
    print("\tThe sugar distribution is statistically different between Salads and Desserts.")
else:
    print("\tThe sugar distribution is not statistically different between Salads and Desserts.")

print('|---------------------------------------------------------------------------------------------|')
```

```python
#  Display both distributions on the same graph for graphical comparison as well

plt.figure(figsize=(10, 6))

sns.kdeplot(salads['Calories'], label='Salads')
sns.kdeplot(desserts['Calories'], label='Desserts')
plt.xlabel('Calories')
plt.ylabel('Density')
plt.title('Distribution of Calories: Salads vs Desserts')
plt.legend()
sns.despine()
plt.tight_layout()
plt.show()
```

8. Is the difference in sugar content between the items in 'Beef & Pork' and 'Chicken and Fish' statistically significant?


```python
from scipy.stats import shapiro
from scipy.stats import mannwhitneyu

# Extracting sugar data for 'Beef & Pork' and 'Chicken & Fish' categories
beef_pork_sugar = merged.loc[merged['Category_x'] == 'beef & pork', 'Sugars']
chicken_fish_sugar = merged.loc[merged['Category_x'] == 'chicken & fish', 'Sugars']

# Testing for normality in the distributions
stat_beef_pork, p_val_beef_pork = shapiro(beef_pork_sugar)
stat_chicken_fish, p_val_chicken_fish = shapiro(chicken_fish_sugar)

# Checking the normality of the distributions and applying the appropriate test
if p_val_beef_pork > 0.5 and p_val_chicken_fish > 0.5:
    print("Both distributions are approximately normal.\n\n")
    t_stat, p_val = ttest_ind(beef_pork_sugar, chicken_fish_sugar)

    if p_val < 0.05:
        print("There is a statistically significant difference in sugar content between 'Beef & Pork' and 'Chicken & Fish'.")
    else:
        print("There is no statistically significant difference in sugar content between 'Beef & Pork' and 'Chicken & Fish'.")
elif p_val_beef_pork > 0.5:
    print("The distribution for 'Chicken & Fish' is approximately normal, but 'Beef & Pork' is not.\n\n")
    t_stat, p_val = mannwhitneyu(beef_pork_sugar, chicken_fish_sugar, alternative='two-sided')

    if p_val < 0.05:
        print("There is a statistically significant difference in sugar content between 'Beef & Pork' and 'Chicken & Fish'.")
    else:
        print("There is no statistically significant difference in sugar content between 'Beef & Pork' and 'Chicken & Fish'.")
elif p_val_chicken_fish > 0.5:
    print("The distribution for 'Beef & Pork' is approximately normal, but 'Chicken & Fish' is not.\n\n")
    t_stat, p_val = mannwhitneyu(beef_pork_sugar, chicken_fish_sugar, alternative='two-sided')

    if p_val < 0.05:
        print("There is a statistically significant difference in sugar content between 'Beef & Pork' and 'Chicken & Fish'.")
    else:
        print("There is no statistically significant difference in sugar content between 'Beef & Pork' and 'Chicken & Fish'.")
else:
    print("Neither of the distributions is approximately normal.\n\n")
    t_stat, p_val = mannwhitneyu(beef_pork_sugar, chicken_fish_sugar, alternative='two-sided')

    if p_val < 0.05:
        print("There is a statistically significant difference in sugar content between 'Beef & Pork' and 'Chicken & Fish'.")
    else:
        print("There is no statistically significant difference in sugar content between 'Beef & Pork' and 'Chicken & Fish'.")
```

10. Does the 'category' factor affect the calorie value? Then, obtain a box-plot graph showing how calorie values are distributed by category.

```python
from scipy.stats import f_oneway

categories = merged['Category_x'].unique()
category_calories = []

for category in categories:
    category_calories.append(merged.loc[merged['Category_x'] == category, 'Calories'])

f_stat, p_val = f_oneway(*category_calories)

if p_val < 0.05:
    print("The 'Category' factor statistically affects the calorie value.")
else:
    print("The 'Category' factor does not statistically affect the calorie value.")


category_sugar = []
for category in categories:
    category_sugar.append(merged.loc[merged['Category_x'] == category, 'Sugars'])

f_stat, p_val = f_oneway(*category_sugar)

if p_val < 0.05:
    print("The 'Category' factor statistically affects the sugar value.")
else:
    print("The 'Category' factor does not statistically affect the sugar value.")

```

```python
#Box-plots

plt.figure(figsize=(10, 6))
plt.boxplot(category_calories, labels=categories)
plt.xlabel('Category')
plt.ylabel('Calories')
plt.title('Calories Distribution by Category')
plt.xticks(fontsize=10, rotation='vertical')
plt.yticks(fontsize=10)
plt.show()


plt.figure(figsize=(10, 6))
plt.boxplot(category_sugar, labels=categories)
plt.xlabel('Category')
plt.ylabel('Sugar')
plt.title('Sugar Distribution by Category')
plt.xticks(fontsize=10, rotation='vertical')
plt.yticks(fontsize=10)
plt.show()
```
